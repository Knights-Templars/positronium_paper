```python
import pandas as pd
import matplotlib.pyplot as plt

%config InlineBackend.figure_format ='retina'
```

### The file for the escape energy contains time (in secs) in the columns, energy (keV) as the rows and the luminosity density (ergs/s/Hz) as values

```python
escape_energy = 'def_n5_100d_500ts_7e7_log_ps1_escape_energy.csv'
```

```python
escape_energy_df = pd.read_csv(escape_energy, sep=',', index_col=0)
```

```python
escape_energy_df
```

```python
times = escape_energy_df.columns.astype('float').values
times_days = times / 86400.0 
escape_energy_df.columns = times_days
```

```python
# select spectra to plot 

spectra_select = escape_energy_df.loc[:, (escape_energy_df.columns > 24) & (escape_energy_df.columns < 24.1)]
```

```python
spectra_select
```

```python
plt.plot(spectra_select)
plt.loglog()
plt.xlabel(r'$\rm Energy~(keV)$')
plt.ylabel(r'$\rm Luminosity~density (erg~s^{-1}~Hz^{-1})$')
plt.title(r'$\rm \gamma-ray~spectra~for~%s~day$' % round(spectra_select.columns.tolist()[0], 2))
```
